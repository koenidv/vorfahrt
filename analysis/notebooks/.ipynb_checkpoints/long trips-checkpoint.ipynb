```python
%load_ext sql

import pandas as pd
import numpy as np
import scipy as sp
import plotly.express as px
```

```python
%sql postgresql://postgres:local@localhost:5432/vorfahrt
vehicles = 0
```

```python
%%sql
vehicles << 
SELECT "milesId", "status", "latitude", "longitude"
FROM "MilesVehicleLastKnown"
WHERE "latitude" < 45 or "latitude" > 55 or "longitude" < 3 or "longitude" > 16
```

```python
df = pd.DataFrame(vehicles)
df
```

```python
# display far-away vehicles

px.scatter_map(df,
                  lat="latitude",
                  lon="longitude",
                  color="status",
                  size_max=15,
                  zoom=10
                  )
```
